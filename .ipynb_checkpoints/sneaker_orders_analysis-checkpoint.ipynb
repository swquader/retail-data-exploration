***Summary***

*The original AOV calculation was incorrect due to the presence of bad data. I was able to find the incorrect data by identifying outliers. After cleaning the dataset, the AOV appears to be $302.58.*

*The metric I would report for this data is the average number of orders per day because it gives us a more intuitive sense of how the business is doing. The average number of orders per day is around 165 orders.*

### User 607

```python
sneaker_orders[sneaker_orders["user_id"] == 607].sort_values(['created_at'])

# all incorrect data associated with user_id 607, always at 4:00:00
# given the unlikelihood of orders being placed exactly at 4 o'clock on the dot, we can confidently say that these lines don't reflect any real orders at all, as opposed to being orders that were recorded incorrectly
# orders marked with user_id 607 likely a system issue
```

```python
sneaker_orders2 = (sneaker_orders.loc[sneaker_orders["user_id"] != 607]).copy() # remove orders by user 607
```

```python
sneaker_orders2[["order_amount","total_items"]].describe()

# maximum order_amount is still unexpected
```

### Additional Outliers

```python
sorted(sneaker_orders2["order_amount"].unique(), reverse=True)
```

```python
order_amount_int = sneaker_orders2["order_amount"].astype(int)
total_items_int = sneaker_orders2["total_items"].astype(int)
```

```python
sneaker_orders2.loc[:, "order_amount"] = order_amount_int
sneaker_orders2.loc[:, "total_items"] = total_items_int
```

```python
sneaker_orders2["price_of_one"] = sneaker_orders2["order_amount"] / sneaker_orders2["total_items"]
```

```python
sneaker_orders2[sneaker_orders2["shop_id"] == 78]

# system likely has the incorrect price logged for the sneaker model sold at shop 78
# while these are likely actual orders, best to remove this data until we can get a better idea of what the true prie is
# alternatively could impute data by estimating the price by looking at the average sneaker price across all other shops
# however, the estimated price wouldn't reflect the specific sneaker model for shop 78
```

```python
sneaker_orders3 = (sneaker_orders2.loc[sneaker_orders2["shop_id"] != 78]).copy()
```

# AOV

```python
sneaker_orders3.describe()[["order_amount","total_items","price_of_one"]]

# The best estimate for the true AOV is $302.58
```

# New Metric

The AOV is hard to interpret without additional context (is $302.58 a good AOV or a bad one?).

A more useful way to understand the business would be to see how much traffic the shops are getting. We can look at this by looking at the daily average of the number of orders. This metric is intuitive and easy to understand.

```python
sneaker_orders3.head()
```

```python
sneaker_orders3["date"] = sneaker_orders3["created_at"].apply(lambda x: pd.to_datetime(x).date())
```

```python
num_orders = sneaker_orders3.groupby("date").count()["order_id"] # series containing total count of orders for each day in March

avg_daily_orders = np.mean(num_orders)

avg_daily_orders
```
